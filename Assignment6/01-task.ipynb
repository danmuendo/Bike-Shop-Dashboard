# Task-01: Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Step 1: Load and Explore the Data


```python
file_path = "Boonsong_Lekagul_waterways_readings.csv"
data = pd.read_csv(file_path)
```

# Display dataset info and first few rows

```python
print("Dataset Information:")
data.info()
print("\nFirst 5 Rows:")
print(data.head())
```

# Check for missing values

```python
print("\nMissing Values:")
print(data.isnull().sum())
```

# Step 2: Analyze the Data
# Trend analysis: Example for "Dissolved oxygen"

```python
oxygen_data = data[data["measure"] == "Dissolved oxygen"]
```

```python
plt.figure(figsize=(10, 6))
oxygen_data.groupby("sample date")["value"].mean().plot()
plt.title("Trend of Dissolved Oxygen Over Time")
plt.xlabel("Sample Date")
plt.ylabel("Average Value")
plt.xticks(rotation=45)
plt.show()
```

# Anomaly detection: Boxplot for all measures

```python
plt.figure(figsize=(10, 6))
data.boxplot(column="value", by="measure", rot=90)
plt.title("Anomalies Across Measures")
plt.ylabel("Value")
plt.show()
```

# Step 3: Select Top 10 Measures
# Analyze the measures in the dataset


```python
print("\nUnique Measures in the Dataset:")
print(data["measure"].unique())
```

### 1a) Trends of Interest

The dataset reveals significant variability in dissolved oxygen levels over time, with values ranging from 6 to 14 mg/L. Such fluctuations suggest potential seasonal patterns or localized environmental conditions. Additionally, the presence of numerous measures, including nutrients (e.g., nitrates) and chemical contaminants (e.g., petroleum hydrocarbons), offers insights into diverse environmental processes. The wide range of measured parameters reflects a comprehensive effort to monitor waterway health across locations.

### 1b) Anomalies in the Dataset

The dataset contains notable anomalies, particularly for measures like "Chemical Oxygen Demand" and "Petroleum Hydrocarbons," with values exceeding 35,000 g/L. These extreme outliers significantly deviate from typical values and may indicate major pollution events or measurement errors. Such anomalies are critical to investigate further, as they could point to industrial discharges, spills, or other environmental hazards.

### 1c) Environmental Impacts

Anomalies and high variability in dissolved oxygen and nutrient levels can disrupt aquatic ecosystems by stressing fish populations and altering biodiversity. Elevated levels of chemical contaminants, such as petroleum hydrocarbons, pose significant risks to water quality and aquatic life, potentially leading to toxic bioaccumulation. These findings highlight the need for targeted environmental interventions and stricter monitoring of potential pollutant sources, especially near industrial areas.

```python
# List all unique measures in the dataset
unique_measures = data['measure'].unique()
print("Unique Measures in the Dataset:")
for measure in unique_measures:
    print(measure)

# Filter and manually select the top 10 relevant measures
selected_measures = [
    "Dissolved oxygen",
    "Chemical Oxygen Demand (Cr)",
    "Ammonium",
    "Nitrates",
    "Total coliforms",
    "Petroleum hydrocarbons",
    "Total phosphorus",
    "Biochemical Oxygen",
    "Fecal streptococci ",
    "PAHs"
]
print("\nSelected Top 10 Measures for Further Analysis:")
print(selected_measures)
```

##### Selected Top 10 Measures:
    - Dissolved oxygen
    - Chemical Oxygen Demand (Cr)
    - Ammonium
    - Nitrates
    - Total coliforms
    - Petroleum hydrocarbons
    - Total phosphorus
    - Biochemical Oxygen
    - Fecal streptococci
    - PAHs (Polycyclic Aromatic Hydrocarbons)


#### Justification for Selection
These measures were chosen based on their critical importance in understanding water quality and identifying pollution sources:

* Dissolved Oxygen: Reflects water's ability to support aquatic life. Fluctuations may signal organic pollution or ecosystem stress.
* Chemical Oxygen Demand (Cr): Indicates the presence of organic pollutants and their impact on oxygen availability.
* Ammonium: High levels can indicate fertilizer runoff or sewage contamination, which are harmful to aquatic ecosystems.
* Nitrates: Contribute to eutrophication, which disrupts water bodies by promoting algal blooms.
* Total Coliforms: Microbial contamination metric; high levels can signal potential health risks.
* Petroleum Hydrocarbons: Measure of industrial pollution, including oil spills, which pose serious environmental threats.
* Total Phosphorus: A key nutrient contributing to eutrophication and the degradation of aquatic environments.
* Biochemical Oxygen: Highlights the amount of oxygen consumed by microbes, often linked to organic pollution.
* Fecal Streptococci: A biological indicator of contamination from human or animal waste.
* PAHs: Toxic compounds often derived from industrial sources, known to be carcinogenic and persistent in the environment.
